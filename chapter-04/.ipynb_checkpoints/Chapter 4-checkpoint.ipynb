```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

import os

PROJECT_PATH = os.path.join('/', 'Users', 'jcaracotsios', 'Projects', 'Textbooks', 'Time-Series-Analysis')
```

```python
def autocorrelation_AR1(k, phi):
    return phi**k

def autocorrelation_AR2(k, phi):
    discriminant = phi[0]**2 + 4*phi[1]
    
    if discriminant < 0: 
        R = np.sqrt(-phi[1])
        Theta = np.arccos(phi[0] / (2*R))
        Phi = np.arctan((1 - phi[1]) / (1 + phi[1]) * np.tan(Theta))
        
        return R**k * np.sin(Theta * k + Phi) / np.sin(Phi)
    elif discriminant > 0:
        G1 = -2*phi[1] / (phi[0] + np.sqrt(phi[0]**2 + 4*phi[1]))
        G2 = -2*phi[1] / (phi[0] - np.sqrt(phi[0]**2 + 4*phi[1]))
        
        num = (1 - G2**2) * G1**(k+1) - (1 - G1**2) * G2**(k+1)
        denom = (G1 - G2) * (1 + G1*G2)
        
        return num/denom
    else: 
        return (1 + k*((1 + phi[1]) / (1 - phi[1]))) * (phi[0]/2)**k

def autocorrelation_MA2(k, theta):
    denom = 1 + theta[0]**2 + theta[1]**2
    
    if k == 1:
        num = -theta[0] + theta[0]*theta[1]
    elif k == 2:
        num = -theta[1]
    else:
        num = 0
        
    return num/denom

def autocorrelation_ARMA11(k, phi, theta):
    num = (1 - theta*phi)*(phi - theta)
    denom = 1 - 2*theta*phi + theta**2
    
    return num/denom*(phi**(k-1))
```

## Examples

(a) $\phi = 0.9$

```python
lags = range(1, 13)
rho = [autocorrelation_AR1(k, phi=0.9) for k in lags]

plt.stem(lags, rho)
p = plt.xticks(lags)
p = plt.ylim(0, 1)
plt.xlabel(r'Lag $k$')
plt.ylabel(r'$\rho_k$')
```

## Exercises

### Problem 4.1: Sketch the autocorrelation functions for each of the following models:
(a) AR(2) with $\phi_1 = 1.2$ and $\phi_2 = -0.7$

(b) AR(2) with $\phi_1 = -1$ and $\phi_2 = -0.6$

(c) MA(2) with $\theta_1 = 1.2$ and $\theta_2 = -0.7$

(d) MA(2) with $\theta_1 = -1$ and $\theta_2 = -0.6$

(e) ARMA(1, 1) with $\phi = 0.7$ and $\theta = 0.4$

(f) ARMA(1, 1) with $\phi = 0.7$ and $\theta = -0.4$

The second-order autoregressive process has autocorrelation function:
$$\rho_k = \left(1 + \frac{1 + \phi_2}{1 - \phi_2}k\right)\left(\frac{\phi_1}{2}\right)^{k}$$

The second-order moving average process has autocorrelation function (for $k = 1, 2$):
$$\rho_k = \frac{-\theta_k + \theta_1\theta_{k+1}}{1 + \theta_1^2 + \theta_2^2}$$

The (1, 1)-order ARMA process has autocorrelation function:
$$\rho_k = \frac{(1 - \theta\phi)(\phi - \theta)}{1 - 2\theta\phi + \theta^2}\phi^{k-1}$$

```python
krange = list(range(1, 11))

plotdata = pd.DataFrame()
plotdata['k'] = krange
plotdata['AR(2; 1.2, -0.7)'] = [autocorrelation_AR2(k, phi=[1.2, -0.7]) for k in krange]
plotdata['AR(2; -1, -0.6)'] = [autocorrelation_AR2(k, phi=[-1, -0.6]) for k in krange]
plotdata['MA(2; 1.2, -0.7)'] = [autocorrelation_MA2(k, theta=[1.2, -0.7]) for k in krange]
plotdata['MA(2; -1, -0.6)'] = [autocorrelation_MA2(k, theta=[-1, -0.6]) for k in krange]
plotdata['ARMA(1, 1; 0.7, 0.4)'] = [autocorrelation_ARMA11(k, phi=0.7, theta=0.4) for k in krange]
plotdata['ARMA(1, 1; 0.7, -0.4)'] = [autocorrelation_ARMA11(k, phi=0.7, theta=-0.4) for k in krange]
```

```python
plt.figure(figsize=(10, 7))
plt.plot(plotdata['k'], plotdata['AR(2; 1.2, -0.7)'], 'o-', label='AR(2; 1.2, 0.7)')
plt.plot(plotdata['k'], plotdata['AR(2; -1, -0.6)'], 'o-', label='AR(2; -1, -0.6)')
plt.plot(plotdata['k'], plotdata['MA(2; 1.2, -0.7)'], 'o-', label='MA(2; 1.2, 0.7)')
plt.plot(plotdata['k'], plotdata['MA(2; -1, -0.6)'], 'o-', label='MA(2; -1, -0.6)')
plt.plot(plotdata['k'], plotdata['ARMA(1, 1; 0.7, 0.4)'], 'o-')
plt.plot(plotdata['k'], plotdata['ARMA(1, 1; 0.7, -0.4)'], 'o-')
plt.legend()
plt.xlabel('Lag (k)')
plt.ylabel('Autocorrelation (rho)')
```
